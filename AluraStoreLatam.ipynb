### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.head()
```

```python
import numpy
import matplotlib.pyplot as plt
import seaborn as sns
```

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
tiendas = [tienda, tienda2, tienda3, tienda4]

for i, t in enumerate(tiendas, start=1):
    total = t["Precio"].sum()
    porcentaje = ((total / sum(t["Precio"].sum() for t in tiendas)) * 100).round()
    print(f"La tienda {i} tiene un ingreso total de: {total}, con porcentaje {porcentaje}%")
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
ingresos = []

for i, t in enumerate(tiendas, start=1):
    total = t["Precio"].sum()
    ingresos.append(total)
    print(f"La tienda {i} tiene un ingreso total de: {total}")

etiquetas = [f"Tienda {i}" for i in range(1, len(tiendas) + 1)]

plt.figure(figsize=(8,5))
sns.barplot(x=etiquetas, y=ingresos, palette="coolwarm", edgecolor="black")
plt.title("Ingresos por Tienda", fontsize=14, fontweight="bold")
plt.ylabel("Ingreso Total ($)", fontsize=12)
plt.xlabel("Tienda", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
tiendas = [tienda, tienda2, tienda3, tienda4]

for i, ct in enumerate(tiendas, start=1):
    catgorias = ct["Categoría del Producto"].value_counts()


    print(f"Categorias más vendidad en la tienda {i}  \n {catgorias}\n")
```

```python
for i, ct in enumerate(tiendas, start=1):
    categorias = ct["Categoría del Producto"].value_counts()

    plt.figure(figsize=(8, 5))
    sns.barplot(x=categorias.values, y=categorias.index, palette="coolwarm", edgecolor="black")
    plt.title(f"Categorías más vendidas - Tienda {i}", fontsize=14, fontweight="bold")
    plt.xlabel("Cantidad vendida", fontsize=12)
    plt.ylabel("Categoría", fontsize=12)
    plt.grid(axis="x", linestyle="--", alpha=0.7)
    plt.tight_layout()
    plt.show()
```

# 3. Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
tienda.columns
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]

for i, cf in enumerate(tiendas, start=1):
     calificaciones = cf["Calificación"].mean()
     calificaciones = calificaciones.round(2)
     print(f"Calificacion de la tienda {i}: \n {calificaciones} \n")
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
calificaciones = []

for cf in tiendas:
    total = cf["Calificación"].mean()
    calificaciones.append(total)
etiquetas = [f"Tienda {i}" for i in range(1, len(tiendas) + 1)]

plt.figure(figsize=(10,5))
sns.barplot(x=etiquetas, y=calificaciones, palette="coolwarm", edgecolor="black")
plt.title("Calificación promedio por tienda", fontsize=14, fontweight="bold")
plt.ylabel("Calificación", fontsize=12)
plt.xlabel("Tienda", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
tienda.columns
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]

for i, ct in enumerate(tiendas, start=1):
    categorias = ct["Producto"].value_counts()
    top_mas_vendidos = categorias.head(3)
    top_menos_vendidos = categorias.tail(3)

    print(f"\nEn la tienda {i}:")
    print("  Top 3 productos más vendidos:")
    for producto, cantidad in top_mas_vendidos.items():
        print(f"    - {producto}: {cantidad} ventas")

    print("  Top 3 productos menos vendidos:")
    for producto, cantidad in top_menos_vendidos.items():
        print(f"    - {producto}: {cantidad} ventas")
```

# 5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

```python
tiendas = [tienda, tienda2, tienda3, tienda4]

for i, cf in enumerate(tiendas, start=1):
     costoEnvio = cf["Costo de envío"].mean()
     print(f"Promedio de costo de envio de la tienda {i}: \n {costoEnvio} \n")
```
